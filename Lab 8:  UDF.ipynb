```python
from pyspark.sql.types import StructType, StructField, StringType

data = [("Ram", "Databricks"),
        ("Dheeraj", "SQL"),
        ("Aman", "Ansible")]

schema = StructType([
    StructField("name", StringType(), True),
    StructField("technology", StringType(), True)
])

df = spark.createDataFrame(data, schema)
display(df)
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType

def string_type(input_str):
    return len(input_str)
```

```python
length_udf = udf(string_type, IntegerType())

df_with_length =  df.withColumn("technology_length", length_udf(df["technology"]))

df_with_length.show()
```

```python
df.createOrReplaceTempView("people")

spark.udf.register("string_length_udf", string_type, IntegerType())

spark.sql("""SELECT name, technology, string_length_udf(technology) as technology_length 
          FROM people
        """).show()
```
