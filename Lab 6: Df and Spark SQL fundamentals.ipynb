```python

df = spark.read.option("delimiter", "\t").csv("s3a://sankethadoop/u.data", inferSchema=True)


df = df.withColumnRenamed("_c0", "user_id") \
    .withColumnRenamed("_c1", "movie_id") \
    .withColumnRenamed("_c2", "rating") \
    .withColumnRenamed("_c3", "timestamp")

df.show()

df.groupBy("rating").count().orderBy("rating").show()
```

```python
# Filter rows where rating is greater than 3
filtered_df = df.filter(df["rating"] > 3)
filtered_df.show(5)

# Select only specific columns: userid and rating
selected_df = df.select("user_id", "rating")
selected_df.show(5)

# Group by movieid and calculate average rating
aggregated_df = df.groupBy("movie_id").agg({"rating": "avg"})
aggregated_df.show(5)
```

```python
df.createOrReplaceTempView("ratings")
```

```python
%sql
SELECT movie_id, AVG(rating) AS avg_rating
FROM ratings
GROUP BY movie_id
ORDER BY avg_rating DESC
LIMIT 5;
```
