**How to Read data from Control plane, DBFS and S3.**

```python
# Read data from Control plane

df = spark.read.option("delimiter", "\t").csv("dbfs:/Workspace/Users/dheeraj.sain@thinknyx.com/data/ds1.data", inferSchema=True)

df.show()
```

```python
df = df.withColumnRenamed("_c0", "user_id") \
    .withColumnRenamed("_c1", "movie_id") \
    .withColumnRenamed("_c2", "rating") \
    .withColumnRenamed("_c3", "timestamp")

df.show()
```

```python
df.groupBy("rating").count().orderBy("rating").show()
```

```python
# Read data from S3

df = spark.read.option("delimiter", "\t").csv("s3a://sankethadoop/u.data", inferSchema=True)


df = df.withColumnRenamed("_c0", "user_id") \
    .withColumnRenamed("_c1", "movie_id") \
    .withColumnRenamed("_c2", "rating") \
    .withColumnRenamed("_c3", "timestamp")

df.show()

df.groupBy("rating").count().orderBy("rating").show()

```
