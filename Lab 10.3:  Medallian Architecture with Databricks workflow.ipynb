```python

# -------------------------------------------
# Step 1: Read output of Notebook 2
# -------------------------------------------

input_path = "dbfs:/FileStore/tables/YOURNAME/processed_data/transformed_batting_data"

df = spark.read.format("delta").load(input_path)
```

```python
# -------------------------------------------
# Step 2: Count total number of rows
# -------------------------------------------

total_rows = df.count()
print("Total number of rows:", total_rows)
```

```python
# -------------------------------------------
# Step 3: Find average runs per year
# -------------------------------------------

from pyspark.sql import functions as F

avg_runs_per_year = (
    df.groupBy("year")
      .agg(F.avg("runs").alias("avg_runs"))
      .orderBy("year")
)

print("Average runs per year:")
avg_runs_per_year.show()
```

```python
# -------------------------------------------
# Step 4: Count distinct teams
# -------------------------------------------

distinct_teams = df.select("team").distinct().count()
print("Number of distinct teams:", distinct_teams)
```
