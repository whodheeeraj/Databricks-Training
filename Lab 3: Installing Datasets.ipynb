```python
df = spark.read.option("delimiter", "\t").csv("dbfs:/Workspace/Users/dheeraj.sain@thinknyx.com/data/ds1.data", inferSchema=True)

df.show(5)
```

```python
print(f"Number of rows: {df.count()}")

print(f"Number of columns: {len(df.columns)}")
```

```python
print("\n")
print(f"----------------------{df.schema}---------------------------------")
print("\n")
print(f"--------------------------{df.printSchema}-----------------------------")
```

```python
df = df.withColumnRenamed("_c0", "user_id") \
    .withColumnRenamed("_c1", "movie_id") \
    .withColumnRenamed("_c2", "rating") \
    .withColumnRenamed("_c3", "timestamp")

df.show(5)

df.describe().show()
```

```python
# Highesh rated movies by finding average ratings per movie

# df.groupBy("movie_id").agg({"rating": "avg"}).orderBy("avg(rating)", ascending=False).show(df.count(),False)

print(df.groupBy("movie_id").avg("rating").orderBy("avg(rating)", ascending=False).show())
```

```python
# identify most famous movie

df.groupBy("movie_id").count().orderBy("count", ascending=False).show()
```

```python
# identify spammers in the dataset
# average ratings per user
df.groupBy("user_id").avg("rating").orderBy("avg(rating)", ascending=False).show(10)
```
