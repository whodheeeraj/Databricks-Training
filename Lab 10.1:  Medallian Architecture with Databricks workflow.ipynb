```python
# -------------------------------------------
# Step 1: Read CSV uploaded to DBFS manually
# -------------------------------------------

# Path where you uploaded the CSV file (example)
input_path = "s3a://sankethadoop/Batting.data"

# Read raw CSV (no header)
df_raw = spark.read.csv(input_path)
```

```python
df = (df_raw.withColumnRenamed("_c0", "playerid")
          .withColumnRenamed("_c1", "year")
        .withColumnRenamed("_c2", "col3")
        .withColumnRenamed("_c3", "team")
        .withColumnRenamed("_c4", "col5")
        .withColumnRenamed("_c5", "col6")
        .withColumnRenamed("_c6", "col7")
        .withColumnRenamed("_c7", "runs")
)
display(df)
```

```python
# -------------------------------------------
# Step 3: Write DataFrame in Delta format
# -------------------------------------------

output_path = "dbfs:/FileStore/tables/YOUR_NAME/processed_data/ingested_batting_data"

df.write.format("delta").mode("overwrite").save(output_path)

print("Data written successfully to:", output_path)
```
