```python
# -------------------------------------------
# Step 1: Read output of Notebook 1
# -------------------------------------------

input_path = "dbfs:/FileStore/tables/YOURNAME/processed_data/ingested_batting_data"

df = spark.read.format("delta").load(input_path)
```

```python
# -------------------------------------------
# Step 2: Filter where year > 2000
# -------------------------------------------

df_filtered = df.filter(df.year > 2000)
```

```python
# -------------------------------------------
# Step 3: Add new column "data_quality" = "Valid"
# -------------------------------------------

from pyspark.sql.functions import lit

df_transformed = df_filtered.withColumn("data_quality", lit("Valid"))

display(df_transformed)
```

```python
# -------------------------------------------
# Step 4: Write it back to DBFS in Delta format
# -------------------------------------------

output_path = "dbfs:/FileStore/tables/YOURNAME/processed_data/transformed_batting_data"

df_transformed.write.format("delta").mode("overwrite").save(output_path)

print("Transformed data written to:", output_path)
```
